# Procedimiento

a. Creación de vectores a y b.

```python
a = np.array([3.1,1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])
```

b. Multiplicación escalar de vectores a y b. ¿Qué se debe hacer para implementar la multiplicación de los vectores?

Para poder hacer la multiplicación escalar, o también llamado producto punto de los dos vectores:

Ambos vectores deben de tener la misma longitud,se multiplica elemento a elemento y luego se efectua la suma de los resultados.

Para realizare esta operación se tiene la función np.dot o el operador @ de numpy.


```python
c = np.dot(a, b)
print(c)
d= a @ b
print (d)
```

c. Multiplicacion punto a punto a a.b

```python
e = a * b 
print(e)
```

d. Construcción matriz A.

```python
A= np.array([(2, -1, -3), (4, 1.5, -2.5), (7.3 , -0.9, 0.2)])
```

e. Transpuesta de A.

```python
f= A.T
print(f)
```

f. Funciones librería Numpy. 

ones --> Devuelve una matriz de forma y tipo determinados, llena de unos.

np.ones(forma, tipo d=Ninguno como float o int, orden='C' para filas ó 'F' para columnas, como=para crear el arreglo con el mismo tipo que otro objeto)




```python
g = np.ones((3,3), int)
print(g)
```

round --> Redondear uniformemente al número dado de decimales.

np.round("a" el arreglo que se quiere modificar, decimales=0 número de deciamales a conservar, fuera= para guardar el resultado en una arreglo ya existente)

```python
h = np.round(A, 0)
print(h)
```

ceil --> Sirve para redondear cada número hacia arriba (al entero mayor más cercano), elemento por elemento.

np.ceil("x" arreglo a modificar, /, fuera=para guardar el resultado en una arreglo ya existente, *, donde=Permite aplicar la función solo en ciertas posiciones ejm x>0, casting='mismo_tipo', orden='K', dtype=Ninguno, subok=Verdadero[, firma])=<ufunc'ceil'>

```python
i = np.ceil(A,None, where=A<0)
print(i)
```
